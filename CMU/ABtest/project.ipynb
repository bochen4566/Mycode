### Introduction

In the highly competitive space of social media advertising, standing out and creating a sense of urgency can be key to driving user engagement and conversion rates. WeChat Moments, akin to a newsfeed, is a fertile ground for such interventions. The proposed AB test aims to enhance the effectiveness of WeChat advertising for game and product promotions by introducing social proof in the form of a tally of prior downloads or purchases numbers under the ads profile section. Social proof is the psychological phenomenon where people assume the actions of others reflect correct behavior for a given situation. Specifically, the study seeks to understand whether such an addition can significantly enhance the effectiveness of ads for games and products by influencing user interactions, click-through rates (CTR), and conversion rates.

The research tests two primary hypotheses:
	Null Hypothesis (H0): Adding or modifying small text prompts in WeChat Moments advertisements, indicating the number of prior downloads for games or purchases for products, will not lead to a significant change in the number of subsequent purchases or downloads.

	Alternative Hypothesis (H1): The inclusion of a small text prompt in WeChat Moments ads, showcasing the number of previous downloads for games or purchases for products, will significantly increase user engagement with the ads (measured through comments and likes), the click-through rate (CTR), and the conversion rate (actual downloads or purchases).

### Treatment and Randomization:

The treatment in this study involves the addition of a small text prompt to the ads on WeChat Moments. For game advertisements, the prompt will state "X people have downloaded this game," whereas for product advertisements, it will state "X people have purchased this item." This addition aims to leverage the psychological phenomenon of social proof, where people are influenced by the actions of others, assuming those actions reflect correct behavior.

The questionnaire design begins with an initial screening question to determine the respondent's preference for product recommendations or game downloads in WeChat Moments ads, thereby ensuring that the ad content is attractive to the target respondents and enhancing the potential effectiveness of the ad. Based on the respondents' answers, they are allocated to one of two main branches: product purchase or game download.

Within each branch, respondents are further randomized into control or test groups to ensure the representativeness of each group of respondents. If respondents clearly indicate a preference for a certain type of ad (game or product) in the screening question, they have a 50% chance of being allocated to the test group and a 50% chance of being allocated to the control group. This random allocation ensures a fair comparison between the two groups while maintaining a balance of respondents' self-selected preferences.

For those who did not clearly indicate a preference in the screening question, they are randomly allocated to one of the four groups (game control group, game test group, product control group, product test group). This step ensures that even without a clear preference, each group receives a sufficient sample size, making the research findings more statistically significant and reliable.

After completing the preference and grouping tests, the questionnaire further collects other personal information from the respondents. The purpose of this step is to comprehensively analyze how the personal background of the respondents affects the test results. In this way, the research can not only evaluate the effectiveness of social proof in advertisements but also gain deeper and more detailed insights into different groups' reactions to such ads.

```python
from statsmodels.formula.api import ols
regression_results = {}
control['treatment_dummy'] = 0
treatment['treatment_dummy'] = 1
switchbackdata = control.append(treatment)
outcome_variables = ['Q1', 'Q2', 'Q3', 'Q4']
for outcome in outcome_variables:
    model = ols(f"{outcome} ~ treatment_dummy", data=switchbackdata).fit()
    regression_results[outcome] = {
        "coef_treat": model.params['treatment_dummy'],
        "p-value_treat": model.pvalues['treatment_dummy'],
        "R-squared": model.rsquared
    }
regression_results
```

```python
# effects with page stay
outcome_variables = ['Q1', 'Q2', 'Q3', 'Q4']
['page_stay', 'page_click', 'usage', 'Age_dummy', 'gender_dummy', 'US', 'Non_US']
# Run simple regression for each outcome variable
for outcome in outcome_variables:
    model = ols(f"{outcome} ~ treatment_dummy + page_stay", data=switchbackdata).fit()
    regression_results[outcome] = {
        "coef_treat": model.params['treatment_dummy'],
        "coef_stay": model.params['page_stay'],
        "p-value_treat": model.pvalues['treatment_dummy'],
         "p-value_stay": model.pvalues['page_stay'],
        "R-squared": model.rsquared
    }
regression_results
```

```python
# effects with usage
outcome_variables = ['Q1', 'Q2', 'Q3', 'Q4']
['page_stay', 'page_click', 'usage', 'Age_dummy', 'gender_dummy', 'US', 'Non_US']
# Run simple regression for each outcome variable
for outcome in outcome_variables:
    model = ols(f"{outcome} ~ treatment_dummy + usage", data=switchbackdata).fit()
    regression_results[outcome] = {
        "coef_treat": model.params['treatment_dummy'],
        "coef_usage": model.params['usage'],
        "p-value_treat": model.pvalues['treatment_dummy'],
         "p-value_usage": model.pvalues['usage'],
        "R-squared": model.rsquared
    }
regression_results
```

### Limitaions

Using a very small sample size in A/B testing like ours, particularly under 60 observations, and relying solely on questionnaires rather than actual user actions on a page, presents several limitations that can significantly impact the reliability, validity, and applicability of the test results. Here are some of the key limitations:

Statistical Power: A small sample size reduces the statistical power of the test, making it harder to detect a true effect if one exists. This means there's a higher risk of Type II errors (failing to reject a false null hypothesis), which can lead to overlooking meaningful differences between the experimental and control groups. Here, the regression analysis can lead to insufficient number of data collected can lead to some misfindings. 

Representativeness: With such a small sample, it's challenging to ensure that the sample is representative of the larger population. This lack of representativeness can limit the generalizability of the findings, meaning the results might not be applicable to a broader audience or different contexts. Here, our groups are only measured by the students, there are lots of social groups using social media not included.

Variability and Noise: Small sample sizes are more susceptible to the effects of random variability and outliers. A few atypical responses can disproportionately affect the results, leading to misleading conclusions about the effectiveness of the tested changes. Here, we want to measure the change due to about 20 observations is not so robust.

Reliance on Self-Reported Data: Using questionnaires to measure outcomes, rather than observing actual user behavior, introduces subjectivity and potential biases into the data. Respondents may not accurately remember their actions or may answer in a socially desirable manner, which can distort the findings. The main idea was to measure how user will interact with this on social media rather than just answering questions.

Limited Insights into User Behavior: Questionnaires can provide valuable insights into users' attitudes and self-reported behaviors, but they lack the depth of understanding that can be gained from observing actual interactions with the page. This can lead to a gap between what people say they do and what they actually do, limiting the ability to make informed decisions.
